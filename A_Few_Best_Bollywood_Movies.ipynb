```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
r=requests.get('https://www.timeout.com/london/film/the-100-best-bollywood-movies-the-list-at-a-glance')
soup=BeautifulSoup(r.text,'lxml')

records=[]
for result in soup.find_all('div',class_='row'):
    title=result.img['alt'][:-7]
    #print title
    director=result.find('div',class_='feature_item__annotation--truncated xs-mb3 slab').contents[2].text[10:]
    #print director
    cast=result.find('div',class_='feature_item__annotation--truncated xs-mb3 slab').contents[4].text[6:]
    #print cast
    genre=result.find('div',class_='feature_item__annotation--truncated xs-mb3 slab').contents[6].text[7:]
    records.append((title,director,cast,genre)) 
df=pd.DataFrame(records,columns=['TITLE','DIRECTOR','CAST','GENRE'])
df.to_csv('A_Few_Best_Bollywood_Movies.csv', index=False, encoding='UTF-8')
pd.read_csv('A_Few_Best_Bollywood_Movies.csv', encoding='UTF-8')
```
